```python
client = pymongo.MongoClient("mongo")
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head(50)
```

### Netoyer les données

```python
#df_youtube = df_youtube.rename(columns = {"video_id": "_id"})
```

```python
document = df_youtube.to_dict(orient = 'records')
```

```python
with open('./data/US_category_id.json') as data:
    category_dict = json.load(data)
    print(category_dict)
    
```

```python
for json1 in document:
    for json2 in category_dict["items"]:
        if json1['category_id'] == int(json2["id"]):
            json1["category_name"] = json2["snippet"]["title"]
            
```
